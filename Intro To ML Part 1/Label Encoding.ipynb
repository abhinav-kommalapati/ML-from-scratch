Label Encoding:
- Converting the lables into numeric form

```python
#importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
#loading the data frfom csv file to pandas dataframe

cancer_data=pd.read_csv('/content/breast_cancer_data.csv')
```

```python
cancer_data.head()
```

```python
#finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
#loading the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to the data frame
cancer_data['target'] = labels
```
