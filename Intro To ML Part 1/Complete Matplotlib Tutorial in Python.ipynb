Matplotlib

- useful for making plots

```python
#importing Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z= np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave

plt.plot(x,z)
plt.show()
```

```python
#adding title x axis and y axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r.') # prints . in red
plt.show()
```

```python
plt.plot(x,y, 'g+') # prints + in green
plt.show()
```

```python
plt.plot(x,y, 'b. ') # prints . in blue
plt.show()
```

```python
#graphing multiple lines

x = np.linspace(-5,5,100)
plt.plot(x,np.sin(x), 'g-')
plt.plot(x,np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #[left, bottom, width, height]
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie Chart

```python
from enum import auto
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['eng','hindi','french','kanada','latin']
people= [ 100, 50, 150, 40,70]
ax.pie(people, labels = languages,autopct='%1.2f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python

fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
