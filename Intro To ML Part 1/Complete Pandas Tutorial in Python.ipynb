Pandas Data Frame:
Pandas DataFrame is a two-dimensional tabular data structure with labled axes(rows & columns)

```python
import pandas as pd
import numpy as np

```

Creating a Pandas DataFrame

```python
#importing the california house price datset


from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing)
```

```python
california_df = pd.DataFrame(housing.data, columns = housing.feature_names)
```

```python
california_df.shape
```

Manipulating a DataFrame

```python
#adding a column to the Data Frame
california_df['price'] = housing.target
```

```python
#removing a row
california_df.drop(index = 0 , axis = 0 )
```

```python
#removing a column

california_df.drop(columns = "Longitude", axis = 1)
```

```python
#locating a row using the index value
california_df.iloc[2]
```

```python
#locating a particular column
print(california_df.iloc[:,0]) #first colums
print(california_df.iloc[:,1])#second column
print(california_df.iloc[:,2])#third column
print(california_df.iloc[:,-1])#last column
```
