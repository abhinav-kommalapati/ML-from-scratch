Imbalanced Dataset:

A dataset with unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
#loading the datset

credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.isnull().sum()
```

```python
#distribution of the 2 classes
credit_card_data['Class'].value_counts()
```

This is highly inbalanced dataset

0 --> Legit Transaction
1 --> Fraud Transactiopn

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

under-sampling

Build a sample dataset containg similar  distribution of legit & fraud transaction


number of fraud transaction -> 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames



```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
