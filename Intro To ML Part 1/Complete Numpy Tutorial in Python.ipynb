List vs NumMPy - Time Taken

```python
from time import process_time
```

Time taken by a list

```python
python_list = [i for i in range(10000)]

start_time = process_time()
python_list = [i + 5 for i in python_list]
end_time = process_time()
print(end_time - start_time)
```

Time taken bys using Numpy arrays

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()
np_array += 5
end_time = process_time()
print(end_time - start_time)
```

Initial Placeholders in numpy arrays

```python
#creating a numpy array of zeros

x = np.zeros((4,5))
print(x)
```

```python
#creating a numpy array of ones
z = np.ones((3,3),dtype= int)
print(z)
```

```python
#arrays of a particular value

z = np.full((5,4),5)
print(z)
```

```python
#creating an idientity matrix
# 4 gives 4x4 matrix

a = np.eye(4)
print(a)
```

```python
# creating a numpy array with random values

b = np.random.rand(4,3)
print(b)
```

```python
#random integer array within a specific range

c = np.random.randint(0,10,(3,5))
print(c)
```

```python
#array of evenly spaced values
#evenly spaced 5 values b/w 10-30

d= np.linspace(10,30,5)
print(d)
```

```python
#array of evenly spaced values --> specifying the step(5)

e = np.arange(10,30,5)
print(e)
```

Mathematical opperations on a numpy array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

print(list1 + list2) # concatinates 2 lists
```

```python
a = np.random.randint(1,10,(3,3))
b = np.random.randint(10,20,(3,3))
print(a)
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python

# reshaping a array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
