```python
# importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import metrics
```

Data Pre-Processing



```python
salary_data = pd.read_csv('/content/salary_data (2).csv')
```

```python


# printing the first 5 columns of the dataframe
salary_data.head()
```

```python

# number of rows & columns in the dataframe
salary_data.shape
```

```python
# checking for missing values
salary_data.isnull().sum()
```

Splitting the feature & target



```python
features = salary_data.iloc[:,:-1].values
target = salary_data.iloc[:,1].values
```

```python
print(features)
```

```python
print(target)
```

Splitting the dataset into training & test data



```python
# Splitting the dataset into training and test data
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import Lasso

# Train the Lasso regression model
model = Lasso(alpha=1.0) # You can adjust alpha as needed
model.fit(features_train, target_train)
```

```python
# Predict the salary values for the test data
test_data_prediction = model.predict(features_test)
```

```python
# Print the test data predictions
print(test_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(target_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(target_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
